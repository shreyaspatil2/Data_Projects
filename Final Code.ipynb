```python
def Working_Team():
    
    #importing important libraries
    import numpy as np
    import pandas as pd
    import datetime
    
    # taking inputs (starting date and number of days)
    print("Enter Starting Date")
    year = int(input("Enter Year: "))
    month = int(input("Enter Month: "))
    day = int(input("Enter Day: "))
    num_days = int(input("Enter Number of Days: "))
    start_date = datetime.date(year, month, day)
    number_of_days = num_days + 50

    # creating list of dates
    date_list = []
    for day in range(number_of_days):
        a_date = (start_date + datetime.timedelta(days = day)).isoformat()
        date_list.append(a_date)
    
    # creating dictionary of dates
    my_dict = dict()
    n = 1
    for i in date_list:
        my_dict.update( {n : i} )
        n=n+1    

    # creating empty dataframe with dates as column names
    df1 = pd.DataFrame(columns = date_list)
    
    # taking number of sites as an input and assigning number of teams of CT,C,E,P
    number_of_sites = int(input("Enter Number of Sites: "))
    n_sites = number_of_sites 
    n_CT = n_C =  15
    n_E = n_P = 7
    
    # assigning working duration of each process
    start_CT = 1
    start_C = start_CT + 7
    start_E = start_C + 5
    start_P = start_E + 2
    
    # filling initial working round in dataframe
    for i in range(1,n_CT+1):
        for j in range(start_CT,start_C):
            df1.loc[i,my_dict[j]] = "CT" + str(i)
        for j in range(start_C,start_E):
            df1.loc[i,my_dict[j]] = "C" + str(i)
        if i <= n_E:
            for j in range(start_E,start_P):
                df1.loc[i,my_dict[j]] = "E" + str(i)
            for j in range(start_P,start_P+2):
                df1.loc[i,my_dict[j]] = "P" + str(i)
    
    # creating lists containing all teams
    CT= []
    C =[]
    E = []
    P = []
    for i in range(1,n_CT+1):
        CT.append("CT"+str(i))
    for i in range(1,n_C+1):
        C.append("C"+str(i))
    for i in range(1,n_E+1):
        E.append("E"+str(i))
    for i in range(1,n_P+1):
        P.append("P"+str(i))             
    x = round(n_sites/n_CT)
    y = round(n_sites/n_E)
    
    # fiiling rest of the dataframe
    for b in range(x):
        for a in CT:
            r, c = np.where(df1 == a)
            value = [r[-1],c[-1]]
            i_row = value[0] + 16
            i_col = value[1] + 1
            for j in range(i_col,i_col+7):
                df1.loc[i_row, date_list[j]] = a
        for a in C:
            r, c = np.where(df1 == a)
            value = [r[-1],c[-1]]
            i_row = value[0] + 16
            i_col = value[1] + 3
            for j in range(i_col,i_col+5):
                df1.loc[i_row, date_list[j]] = a
    for b in range(y):
        for a in E:
            r, c = np.where(df1 == a)
            value = [r[-1],c[-1]]
            i_row = value[0] + 8
            i_col = value[1] + 2
            for j in range(i_col,i_col+2):
                if pd.notnull(df1.loc[i_row, date_list[j]]):
                    df1.loc[i_row, date_list[j+1]] = a
                else:
                    df1.loc[i_row, date_list[j]] = a
        for a in P:
            r, c = np.where(df1 == a)
            value = [r[-1],c[-1]]
            i_row = value[0] + 8
            i_col = value[1] + 1
            for j in range(i_col,i_col+2):
                if pd.notnull(df1.loc[i_row, date_list[j]]):
                    if pd.notnull(df1.loc[i_row, date_list[j+1]]):
                        df1.loc[i_row, date_list[j+2]] = a
                    else: 
                        df1.loc[i_row, date_list[j+1]] = a
                else:
                    df1.loc[i_row, date_list[j]] = a
        
    Site_Number = int(input("\nEnter Required Site Number: "))
    Working_Date = input("Enter Required Date In Format (YYYY-MM-DD): ")
    SSS = df1.loc[Site_Number,Working_Date]
    print("\nCurrent Working Team At Site ", Site_Number," On Day ",Working_Date," is ",SSS)
```

```python
Working_Team()
```
