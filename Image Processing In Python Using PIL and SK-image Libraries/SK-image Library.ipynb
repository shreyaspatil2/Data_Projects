# Scikit Image Library
scikit-image (formerly scikits.image) is an open-source image processing library for the Python programming language.It includes algorithms for segmentation, geometric transformations, color space manipulation, analysis, filtering, morphology, feature detection, and more. It is designed to interoperate with the Python numerical and scientific libraries NumPy and SciPy.

```python
import skimage
dir(skimage)
```

##### Sk-Image Key Points:
* skimage images are stored as three-dimensional NumPy arrays.

* In skimage images, the red channel is specified first, then the green, then the blue, i.e. RGB.

* Images are read from disk with the skimage.io.imread() function.

* We create a window that automatically scales the displayed image with skimage.viewer.ImageViewer() and calling view() on the viewer object.

* Color images can be transformed to grayscale using skimage.color.rgb2gray() or be read as grayscale directly by passing the argument as_gray=True to skimage.io.imread().

* We can resize images with the skimage.transform.resize() function.

* NumPy array commands, like img[img < 128] = 0, and be used to manipulate the pixels of an image.

* Array slicing can be used to extract sub-images or modify areas of images, e.g., clip = img[60:150, 135:480, :].

* Metadata is not retained when images are loaded as skimage images.

# 5. Convert RGB to HSV 

#### Other two popular formats are HSV (hue, saturation, value) and HSL (hue, saturation, lightness) which are alternative representations of the RGB format. Let me briefly explain what each of these terms mean.

* Hue is a degree on the color wheel where 0 is for red, 120 is green, 240 is blue and again 360 would be red.
* Saturation represents the percentage of that color, where 0 is white and 100  is the full color.
* Value denotes the mixture of the colors with varying amounts of black or white paint.
* Lightness is another way to show the shade of the image where 0 is black and 1 is white.


```python
from skimage.color import rgb2hsv
rgb_img = img
hsv_img = rgb2hsv(rgb_img)
imshow(hsv_img)
```

# 7. Pyramiding

```python
import numpy as np
import matplotlib.pyplot as plt

from skimage import data
from skimage.transform import pyramid_gaussian


image = data.astronaut()
#img = imread("wallpapertip_attack-on-titan-wallpaper_10287.jpg")
#k=np.array(img.shape[:2]).min()
#image = resize(img,(k,k))
#print(image)
rows, cols, dim = image.shape
pyramid = tuple(pyramid_gaussian(image, downscale=2, multichannel=True))

composite_image = np.zeros((rows, cols + cols // 2, 3), dtype=np.double)

composite_image[:rows, :cols, :] = pyramid[0]

i_row = 0
for p in pyramid[1:]:
    n_rows, n_cols = p.shape[:2]
    composite_image[i_row:i_row + n_rows, cols:cols + n_cols] = p
    i_row += n_rows

fig, ax = plt.subplots()
ax.imshow(composite_image)
plt.show()
skimage.io.imsave("pyramid.png",composite_image)
```

# 8. Resize

```python
from skimage.transform import resize
img = imread('virat.jpg')
#resize image
img_resized = resize(img, (300, 300))

#plot images
plt.subplot(121), imshow(img)
plt.title('Original Image')
plt.subplot(122), imshow(img_resized)
plt.title('Resized Image')
plt.show()
```

# 9. Rescale

```python
from skimage.transform import rescale
img = imread('virat.jpg')
img_rescaled = rescale(img, scale=(0.5,0.5))


plt.subplot(121), imshow(img)
plt.title('Original Image')

plt.subplot(122), imshow(img_rescaled)
plt.title('Rescaled Image')

plt.show()
```

# 10. Rotate

```python
from skimage.transform import rotate
image = imread('virat.jpg')

image_rotated1 = rotate(image, angle=-30, resize=False)
image_rotated2 = rotate(image, angle=-30, resize=True)

plt.subplot(121), imshow(image_rotated1)
plt.subplot(122), imshow(image_rotated2)

plt.show()
```

# 13. Adjust Exposure

```python
from skimage import exposure

#adjusting brightness
image = imread('virat.jpg')
image_bright = exposure.adjust_gamma(image, gamma=0.5,gain=1)
image_dark = exposure.adjust_gamma(image, gamma=3,gain=1)

# plotting images
plt.figure(figsize=(20,20))
plt.subplot(131), imshow(image)
plt.title('Original Image')

plt.subplot(132),imshow(image_bright)
plt.title('Bright Image')

plt.subplot(133),imshow(image_dark)
plt.title('Dark Image')

plt.show()
```

# 14. Filters

```python
from skimage.filters import median

image = imread('virat.jpg', as_gray=True)
image_median = median(image)

# plotting images
plt.figure(figsize=(10,10))
plt.subplot(121), imshow(image)
plt.title('Original Image')

plt.subplot(122),imshow(image_median)
plt.title('Smooth Image')

plt.show()
```

```python
from skimage.filters import sobel_h
from skimage.filters import sobel_v

image = imread('virat.jpg', as_gray=True)
image_sobelh = sobel_h(image)
image = imread('virat.jpg', as_gray=True)
image_sobelv = sobel_v(image)

# plotting images
plt.figure(figsize=(10,10))
plt.subplot(121), imshow(image_sobelh,cmap="gray")
plt.title('Horizontal Edge')

plt.subplot(122),imshow(image_sobelv,cmap="gray")
plt.title('Vertical Edge')

plt.show()
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

from skimage.draw import (line, polygon,
                          circle_perimeter,
                          ellipse, ellipse_perimeter,
                          bezier_curve)


fig, ax1 = plt.subplots(ncols=1, nrows=1, figsize=(10, 6))


img = np.zeros((500, 500, 3), dtype=np.double)

# draw line
rr, cc = line(120, 123, 20, 400)
img[rr, cc, 0] = 255

# fill polygon
poly = np.array((
    (300, 300),
    (480, 320),
    (380, 430),
    (220, 590),
    (300, 300),
))
rr, cc = polygon(poly[:, 0], poly[:, 1], img.shape)
img[rr, cc, 1] = 1


# fill ellipse
rr, cc = ellipse(300, 300, 100, 200, img.shape)
img[rr, cc, 2] = 1

# circle
rr, cc = circle_perimeter(120, 400, 15)
img[rr, cc, :] = (1, 0, 0)

# Bezier curve
rr, cc = bezier_curve(70, 100, 10, 10, 150, 100, 1)
img[rr, cc, :] = (1, 0, 0)

# ellipses
rr, cc = ellipse_perimeter(120, 400, 60, 20, orientation=math.pi / 4.)
img[rr, cc, :] = (1, 0, 1)
rr, cc = ellipse_perimeter(120, 400, 60, 20, orientation=-math.pi / 4.)
img[rr, cc, :] = (0, 0, 1)
rr, cc = ellipse_perimeter(120, 400, 60, 20, orientation=math.pi / 2.)
img[rr, cc, :] = (1, 1, 1)

ax1.imshow(img)
ax1.axis('off')
```
